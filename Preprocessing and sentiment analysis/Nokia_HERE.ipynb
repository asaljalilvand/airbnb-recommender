```python
import requests
import pandas as pd

app_id = """mA5T109RcMDuxtTxmyOL"""
app_code = """H5qV_HC8G0UmPiXrhBQWRg"""

listings=pd.read_csv('listings.csv')
attraction=pd.read_csv('Melbourn_Attraction.csv',header=None,names=['Attractions', 'Latitude', 'Longitude'])
print(listings.columns)
print(attraction.columns)
```

```python
att_cordinates = pd.concat([attraction.Latitude,attraction.Longitude],axis=1)
list_cordinates = pd.concat([listings.latitude,listings.longitude],axis=1)
print(att_cordinates)
```

```python
lst_cord=[]
for i in range(len(list_cordinates.latitude.tolist())):
    lst_cord.append(str(list_cordinates.latitude.tolist()[i])+','+str(list_cordinates.longitude.tolist()[i]))
list_cordinates['cord']=pd.Series(lst_cord)
#print(list_cordinates)
att_cord=[]
for i in range(len(att_cordinates.Latitude.tolist())):
    att_cord.append(str(att_cordinates.Latitude.tolist()[i])+','+str(att_cordinates.Longitude.tolist()[i]))
att_cordinates['cord']=pd.Series(att_cord)
print(att_cordinates)
```

```python
for name in attraction.Attractions:
    listings[name]=0
```

```python
transport_mode_type=['car', 'pedestrian', 'publicTransport', 'bicycle', 'truck']
traffic = ['disabled', 'enabled', 'default']

NoRouteFound_count=0

for i in range(len(list_cordinates.cord.tolist())):
    if i%10==0:
        print(int(i*10/len(list_cordinates.cord.tolist()))*'='+'> %'+str(i*100/len(list_cordinates.cord.tolist()))[0:4])
        print('Number of NoRouteFound:',NoRouteFound_count)
    start_point=list_cordinates.cord[i]
    for j in range(len(att_cordinates.cord.tolist())):
        end_point=att_cordinates.cord[j]
        request_api_urL = """https://route.api.here.com/routing/7.2/calculateroute.json?app_id={}&app_code={}&waypoint0={}&waypoint1={}&mode=fastest;{};traffic:{}""".format(app_id, app_code, start_point, end_point,transport_mode_type[2],traffic[1])
        response = requests.get(request_api_urL)
        response_json = response.json()
        try:
            listings[attraction.Attractions[j]][i]=response_json['response']['route'][0]['summary']['travelTime']
        except:
            listings[attraction.Attractions[j]][i]=-1
            NoRouteFound_count+=1

print('Number of NoRouteFound:',NoRouteFound_count)        
        
```
